```python
import tensorflow as tf
import os.path as osp
from analysis import data
import pandas as pd

pred_data_path = osp.join(osp.dirname(data.__file__), "pred_consumption_performance.csv")
df_pred_performance = pd.read_csv(pred_data_path)
col_names = ["agent_id",
            "action",
            "electrical_storage_soc",
            "net_electricity_consumption",
            "non_shiftable_load",
            "solar_generation",
            "carbon_intensity",
            "electricity_pricing"]

df_pred_performance.columns = col_names
df_pred_performance
```

```python

pred_data_path = osp.join(osp.dirname(data.__file__), "pred_consumption_predictions.csv")
df_predictions = pd.read_csv(pred_data_path)

prediction_col_names = ["predicted_load", "predicted_solar"]
df_predictions.columns = prediction_col_names
df_predictions
```

```python
historic_load_names = []
for i in range(10):
    historic_load_names.append(f"load_{i + 1}hours_ago")

# df_historic_loads.columns = historic_load_names
# all_loads = df_pred_performance["non_shiftable_load"]


num_historic_loads = 10

for i in range(1, num_historic_loads):
    shifted = df_pred_performance["non_shiftable_load"].shift(periods=i, fill_value=0.0)
    df_pred_performance[f"load_{i}hours_ago"] = shifted

df_pred_performance

    # df_pred["non_shiftable_load"].shift(periods=1, fill_value=0.0) - df_pred["predicted_load"]
#
# all_loads
```

```python
historic_col_names = []
for i in range(9):
    historic_col_names.append(f"load_{i + 1}hours_ago")

df_load_slopes_08 = (df_pred_performance[historic_col_names[0]] - df_pred_performance[historic_col_names[8]]) / 9
df_load_slopes_03 = (df_pred_performance[historic_col_names[0]] - df_pred_performance[historic_col_names[3]]) / 4
slope_names = ["load_slope_1_10", "load_slope_1_4"]
```

```python
df_pred = pd.concat([df_pred_performance, df_predictions, df_load_slopes_08, df_load_slopes_03], axis=1)
df_pred.columns = col_names + prediction_col_names + slope_names
df_pred
```

```python
historic_consumptions_data_path = osp.join(osp.dirname(data.__file__), "pred_historic_consumptions.csv")
df_historic_consumptions = pd.read_csv(historic_consumptions_data_path)

historic_col_names = []
for i in range(10):
    historic_col_names.append(f"consumption_{i + 1}hours_ago")

df_historic_consumptions.columns = historic_col_names
df_historic_consumptions
```

```python
df_slopes_09 = (df_historic_consumptions[historic_col_names[0]] - df_historic_consumptions[historic_col_names[9]]) / 10
df_slopes_03 = (df_historic_consumptions[historic_col_names[0]] - df_historic_consumptions[historic_col_names[3]]) / 4
slope_names = ["slope_1_10", "slope_1_4"]
```

```python
df_pred = pd.concat([df_pred_performance, df_predictions, df_historic_consumptions, df_slopes_09, df_slopes_03], axis=1)
df_pred.columns = col_names + prediction_col_names + historic_col_names + slope_names
df_pred
```

```python
df_pred["load_error"] = df_pred["non_shiftable_load"].shift(periods=1, fill_value=0.0) - df_pred["predicted_load"]
df_pred.head()
```

```python
train_buildings = [0, 1, 2, 4]
validation_buildings = [3]

df_train_buildings = df_pred.loc[df_pred["agent_id"].isin(train_buildings)]
df_val_buildings = df_pred.loc[df_pred["agent_id"].isin(validation_buildings)]
```

```python
include_columns = [
                    # "electrical_storage_soc",
                    # "action",
                    # "net_electricity_consumption",
                    "non_shiftable_load",
                    # "solar_generation",
                    # "slope_1_10",
                    "slope_1_4",
                    "load_error",
                    # "consumption_1hours_ago",
                    # "consumption_3hours_ago",
                    # "consumption_5hours_ago",
                    # "consumption_7hours_ago"
                    ]

df_train_buildings = df_train_buildings[[x for x in df_train_buildings.columns if x in include_columns]]
df_val_buildings = df_val_buildings[[x for x in df_val_buildings.columns if x in include_columns]]
df_train_buildings
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(32, activation='relu'),
  # tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(32, activation='relu'),
  # tf.keras.layers.Dropout(0.2),
  # tf.keras.layers.Dense(16, activation='relu'),
  # tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.losses.MeanSquaredError(),
              metrics=['mean_absolute_error'])
y_train = df_train_buildings["load_error"]
X_train = df_train_buildings.drop(["load_error"], axis=1)

y_test = df_val_buildings["load_error"]
X_test = df_val_buildings.drop(["load_error"], axis=1)
history = model.fit(x=X_train, y=y_train, batch_size=256, epochs=100000, validation_data=(X_test, y_test))
```
